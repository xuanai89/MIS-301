### **Standardizing column names in df_jobs**

```python
# Standardize column names by stripping whitespace, converting to lowercase, and replacing spaces with underscores
df_jobs.columns = df_jobs.columns.str.strip().str.lower().str.replace(' ', '_')
```

### **Removing duplicate rows from df_jobs**

```python
df_jobs.duplicated().sum()
```

```python
# Remove duplicate in df_jobs
df_jobs.drop_duplicates(inplace=True)
```

### **Converting column to correct type**




```python
# Convert 'salary_usd' to numeric
df_jobs['salary_usd'] = pd.to_numeric(df_jobs['salary_usd'], errors='coerce')
```

```python

# Current exchange rates (as of August 2025)
# Note: These are example rates. For a real application, you would use an API to get live rates.
# (Tỷ Giá Chuyển Đổi Euro Sang Đô-la Mỹ. Đổi Tiền EUR/USD - Wise, n.d.)
eur_to_usd_rate = 1.16
# (Tỷ Giá Chuyển Đổi 1 Nghìn Bảng Anh Sang Đô-la Mỹ. Đổi Tiền GBP/USD - Wise, n.d.)
gbp_to_usd_rate = 1.34

# Create a function to convert salaries to USD
def convert_to_usd(row):
    # Check if salary_usd is a valid number, if not, return NaN
    if pd.isna(row['salary_usd']):
        return np.nan

    if row['salary_currency'] == 'EUR':
        return row['salary_usd'] * eur_to_usd_rate
    elif row['salary_currency'] == 'GBP':
        return row['salary_usd'] * gbp_to_usd_rate
    else:
        # Assumes the salary is already in USD if the currency is not EUR or GBP
        return row['salary_usd']

df_jobs['salary_in_usd_converted'] = df_jobs.apply(convert_to_usd, axis=1)
df_jobs['salary_usd'] = df_jobs['salary_in_usd_converted']

# To keep the dataset clean, we can now set the currency for all to USD
df_jobs['salary_currency'] = 'USD'

# Drop the temporary conversion column
df_jobs = df_jobs.drop(columns=['salary_in_usd_converted'])


```

```python
# Convert 'posting_date' and 'application_deadline' to datetime objects
# The `errors='coerce'` argument will turn any unparseable dates into NaT (Not a Time)
df_jobs['posting_date'] = pd.to_datetime(df_jobs['posting_date'], format='%m/%d/%Y', errors='coerce')
df_jobs['application_deadline'] = pd.to_datetime(df_jobs['application_deadline'], format='%m/%d/%Y', errors='coerce')
```

### **Handling missing values by dropping rows with missing values in key columns and filling in missing values with 'Unknown' or the mode/median for other columns**

```python
# Drop rows with missing values in key columns
df_jobs.dropna(subset=['job_title', 'company_name', 'salary_usd', 'education_required'], inplace=True)
```

```python
# Fill missing values with 'Unknown' for specified columns
df_jobs['required_skills'] = df_jobs['required_skills'].fillna('Unknown')
df_jobs['employee_residence'] = df_jobs['employee_residence'].fillna('Unknown')
df_jobs['experience_level'] = df_jobs['experience_level'].fillna('Not specified')
df_jobs['industry'] = df_jobs['industry'].fillna('Unknown')
df_jobs['company_location'] = df_jobs['company_location'].fillna('Unknown')
```

```python
# Fill missing values in specified columns with the mode of each column
df_jobs['company_size'] = df_jobs['company_size'].fillna(df_jobs['company_size'].mode()[0])
df_jobs['employment_type'] = df_jobs['employment_type'].fillna(df_jobs['employment_type'].mode()[0])
df_jobs['salary_currency']= df_jobs['salary_currency'].fillna(df_jobs['salary_currency'].mode()[0])
```

```python
# Fill missing numerical values with the median of each column
df_jobs['remote_ratio'] = df_jobs['remote_ratio'].fillna(df_jobs['remote_ratio'].median())
df_jobs['years_experience'] = df_jobs['years_experience'].fillna(df_jobs['years_experience'].median())
df_jobs['benefits_score'] = df_jobs['benefits_score'].fillna(df_jobs['benefits_score'].median())
df_jobs['job_description_length'] = df_jobs['job_description_length'].fillna(df_jobs['job_description_length'].median())
```

```python
# Calculates the median of the posting_date column
median_posting_date = df_jobs['posting_date'].median()
# Fills missing values in the posting_date column with the calculated median posting date.
df_jobs['posting_date'] = df_jobs['posting_date'].fillna(median_posting_date)
```

```python
# Fills missing values in the application_deadline column with median
df_jobs['application_deadline'] = df_jobs['application_deadline'].fillna(median_posting_date + pd.Timedelta(days=30))
```

```python
# Calculates the number of missing values (null or NaN) in each column of the df_jobs DataFrame to verify that missing values have been handled
df_jobs.isnull().sum()
```

### **The 4th objective: Develop Employer Attractiveness Score**

```python
df_jobs['salary_usd_scaled'] = (df_jobs['salary_usd'] - df_jobs['salary_usd'].min()) / (df_jobs['salary_usd'].max() - df_jobs['salary_usd'].min())
df_jobs['benefits_score_scaled'] = (df_jobs['benefits_score'] - df_jobs['benefits_score'].min()) / (df_jobs['benefits_score'].max() - df_jobs['benefits_score'].min())
df_jobs['remote_ratio_scaled'] = (df_jobs['remote_ratio'] - df_jobs['remote_ratio'].min()) / (df_jobs['remote_ratio'].max() - df_jobs['remote_ratio'].min())
```

```python
df_jobs['attractiveness_score'] = (
    0.3 * df_jobs['salary_usd_scaled'] +
    0.3 * df_jobs['benefits_score_scaled'] +
    0.4 * df_jobs['remote_ratio_scaled']
)
df_jobs['attractiveness_rank'] = df_jobs.groupby('company_location')['attractiveness_score'].rank(ascending=False)

df_top5 = df_jobs.sort_values(by='attractiveness_score', ascending=False).head(5)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_top5,
    x='attractiveness_score',
    y='company_name',
    hue='company_location'
)

plt.xlim(0.8, 1)
plt.title('Top 5 Most Attractive Employers by Location')
plt.xlabel('Attractiveness Score')
plt.ylabel('Company Name')
plt.legend(title='Location')
plt.tight_layout()
plt.show()
```

### **The 5th objective: Quantify Salary Drivers**

```python
avg_salary = df_jobs.groupby('years_experience', as_index=False)['salary_usd'].mean()

plt.figure(figsize=(8, 6))
sns.lineplot(
    data=avg_salary,
    x='years_experience',
    y='salary_usd',
    marker='o'
)
plt.title("Average Salary by Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Average Salary (USD)")
plt.tight_layout()
plt.show()
```
